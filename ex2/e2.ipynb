```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt


import math
from __future__ import division

data=np.loadtxt(open("applesOranges.csv", "rb"), delimiter=",",skiprows=1)
```

Neuron class

```python
class neuron:

    def sign(self,h):
        if(h>=0):
            return 1
        else: 
            return 0
    
    def setbias(self,b):
        self.bias=b
    def setws(self,ws):   
        self.ws=ws
    def setinp(self,inp):   
        self.inp=inp
    
    def output(self):
        
        return self.sign( (self.ws[0]*self.inp[0])+(self.ws[1]*self.inp[1]) - self.bias)
        
```
